```
df = pd.read_csv('/content/drive/My Drive/Data for Python/cleaned_311_data_hw3.csv', header='infer')
```

```
def check_graffiti(row):

  if 'graffiti'in row.complaint_type.lower():
    return True
  if pd.notnull(row.descriptor) and 'graffiti' in row.descriptor.lower():
    return True
  return False
```

```
df ['graffiti_flag']= df.apply(check_graffiti, axis=1)
```

```
df.groupby('graffiti_flag').count()
```

```
df[(df.graffiti_flag== True)].groupby ('complaint_type').count()
```
